```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='speeding', y='total', data=df)
plt.title('Total Accidents vs. Speeding')
plt.xlabel('Speeding (%)')
plt.ylabel('Total Crashes (per billion miles)')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='alcohol', y='total', data=df, color='orange')
plt.title('Total Accidents vs. Alcohol Involvement')
plt.xlabel('Alcohol Involvement (%)')
plt.ylabel('Total Crashes (per billion miles)')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='ins_premium', y='ins_losses', data=df, color='green')
plt.title('Insurance Premiums vs. Insurance Losses')
plt.xlabel('Insurance Premium ($)')
plt.ylabel('Insurance Losses ($)')
plt.show()
```
